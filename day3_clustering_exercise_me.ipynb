```python
# center and scale the data, create X_scaled
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### USE SCALED DATA TO FIT 2, 3, 4,5,6 clusters

```python
# K-means with 2 clusters on scaled data
kmeans_2s = KMeans(n_clusters=2, n_init=1, random_state=1)
kmeans_2s.fit(X_scaled)
```

```python
# K-means with 3 clusters on scaled data
kmeans_3s = KMeans(n_clusters=3, n_init=1, random_state=1)
kmeans_3s.fit(X_scaled)
```

```python
# K-means with 4 clusters on scaled data
kmeans_4s = KMeans(n_clusters=4, n_init=1, random_state=1)
kmeans_4s.fit(X_scaled)
```

```python
# K-means with 5 clusters on scaled data
kmeans_5s = KMeans(n_clusters=5, n_init=1, random_state=1)
kmeans_5s.fit(X_scaled)
```

```python
# K-means with 6 clusters on scaled data
kmeans_6s = KMeans(n_clusters=6, n_init=1, random_state=1)
kmeans_6s.fit(X_scaled)
```

```python
# Create a pandas dataframe named as df_X to hold the scaled X array data, and name the two columns as 'x' and 'y'
df_X = pd.DataFrame(X_scaled,columns = ['x','y'])

# Add a column named cluster to df_X to hold the kmeans_3s's labels_ attribute
df_X['cluster'] = kmeans_3s.labels_

# Sort df_X by cluster column
df_X.sort_values('cluster')

# Group by df_X by cluster and compute its mean
df_X.groupby('cluster').mean()
```

## EVALUATING K-means

```python
# print inertia for kmeans_2s, 3s,4s,5s,6s
print ("Inertia for KMeans with 2 clusters = %lf " % kmeans_2s.inertia_)
print ("Inertia for KMeans with 3 clusters =  %lf " % kmeans_3s.inertia_)
print ("Inertia for KMeans with 4 clusters = %lf " % kmeans_4s.inertia_)
print ("Inertia for KMeans with 5 clusters = %lf " % kmeans_5s.inertia_)
print ("Inertia for KMeans with 6 clusters =  %lf " % kmeans_6s.inertia_)
```

```python
from sklearn import metrics
# print Silhouette score for kmeans_2s, 3s,4s,5s,6s, use sample size of 200
```

```python
# Which KMeans is the best according to the Silhouette score? 
```

```python
# calculate Silhouette score for K=2 through K=19
k_range = range(2, 20)
scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(X_scaled)
    scores.append(metrics.silhouette_score(X_scaled, km.labels_))
```

```python
# plot the results of Silhouette score from k = 2 to 19
plt.plot(k_range, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.grid(True)
```
